```python
import nltk
nltk.download('punkt')
from nltk import word_tokenize

import nltk
nltk.download('stopwords')
stop = set(nltk.corpus.stopwords.words("arabic"))

def stopRemoval(text):
    temptweet = word_tokenize(text)
    text = " ".join([w for w in temptweet if not w in stop and len(w) >= 2])
    return text

df['noStop_tweet'] = df['text'].apply(lambda x:stopRemoval(x))
df
```

```python

def normalize(text):
    pattern = re.compile(r"(.)\1{2,}", re.DOTALL) # حذف ثلاثة حروف متكررة ورا بعض
    text=pattern.sub(r"\1\1", text)
    text = text.replace(u"\u0625", u"\u0627")  # همزة الى الف
    text = text.replace(u"\u0622", u"\u0627")  # أ ممدودة الى ا
    text = text.replace(u"\u0623", u"\u0627")  # أ الى ا

    text=text.replace(u"\u0629", u"\u0647") # ة الى ه
    
    text = text.replace(u"\u064B", "")  #  تنوين الفتح
    text = text.replace(u"\u064C", "")  #  تنوين الضم
    text = text.replace(u"\u064D", "")  #  تنوين الكسر
    text = text.replace(u"\u064E", "")  #  فتحة
    text = text.replace(u"\u064F", "")  #  ضمة
    text = text.replace(u"\u0650", "")  #  كسرة 
    text = text.replace(u"\u0651", "")  #  شدة 
    text = text.replace(u"\u0652", "")  #  سكون
    text = text.replace(u"\u0670", "`")  #  الحركة اللي فوق الالف  
    return text

df['normalized_tweet'] = df['text'].apply(lambda x: normalize(x))

df['text1'] = df['noStop_tweet'].apply(lambda x:normalize(x))

df.head()
```

```python

import string
import re

arabicPunctuations = [".","`","؛","<",">","(",")","*","&","^","%","]","[",",","ـ","،","/",":","؟",".","'","{","}","~","|","!","”","…","“","–"] # defining customized punctuation marks
englishPunctuations = string.punctuation
englishPunctuations = [word.strip() for word in englishPunctuations]
punctuationsList = arabicPunctuations + englishPunctuations 

def removingPunctuation(text):
  cleanTweet = ''
  for i in text:
    if i not in arabicPunctuations:
      cleanTweet = cleanTweet + '' + i
  return cleanTweet

df['noPunc_tweet'] = df['text'].apply(lambda x: removingPunctuation(x))
df['text1'] = df['text1'].apply(lambda x: removingPunctuation(x))
```

```python
# Remove english
def remove_english_letters(text):
    pattern = re.compile(r"[a-zA-Z]")
    return re.sub(pattern, "", text)

def remove_english_letters_df(text):
    text['cleaned_text1'] = text['cleaned_text1'].apply(remove_english_letters)
    return text
```
