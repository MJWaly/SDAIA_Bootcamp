```python
import pandas as pd
import numpy as np

# Assuming you have separate DataFrames data0, data1, data2, data3, and data4 for each category

# Create a list of DataFrames for each category
category_data = [data0, data1, data2, data3, data4,data5,data6,data7,data8]

# Determine the target balanced count (e.g., 13736)
target_count = data8['label'].count()

balanced_data = []

for i, data in enumerate(category_data):
    # Randomly shuffle the data within each category
    shuffled_data = data.sample(frac=1, random_state=42)

    # Select a subset of data to match the target count
    subset_data = shuffled_data.head(target_count)

    # Save the balanced subset as new DataFrames
    new_data_name = f'data{i}{i}'
    globals()[new_data_name] = subset_data
    
    balanced_data.append(subset_data)

# Concatenate the balanced data from all categories
balanced_dataset = pd.concat(balanced_data, ignore_index=True)
```

```python
def tokenization(x,idx): # take data copy
        tokens = nltk.word_tokenize(x['clean_text'][idx])
        return tokens
    
def stemming(x):
    st = nltk.ISRIStemmer()
    stemming_root =[st.stem(token) for token in x]
    return stemming_root  
```

```python
bd['puretext'] =0

for i in range(0,bd.shape[0]):
    tokens = tokenization(bd , i )  
    pure_tokens = stemming(tokens)
    bd['puretext'][i] = ' '.join(pure_tokens)
```
